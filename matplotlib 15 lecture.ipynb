```python
sal=[15,22,24,25,30,23,34,43,19,20,52,21]
sal.sort()
sal
```

```python
# Outliers
```

```python
import pandas as pd
```

```python
ser=pd.Series(sal)
ser
```

```python
# please see below example
# [15, 19, 20, 21, 22, 23, 24, 25, 30, 34, 43, 52]
# n=total element in the above example (12 item)

# 25% (25/100)*(n+1)=3.25 (data at 3.25 position is the 25% percentile lets say around 20.50)

# 50% (50/100)*(n+1)=6.5 (data at this position)

# 75% (75/100)*(n+1)=9.75 (data at this postion)
```

```python
# 1st quantile
```

```python
q1=ser.quantile(0.25)
print('q1',q1)
```

```python
# 3rd quantile
```

```python
q3=ser.quantile(0.75)
print('q3',q3)
```

```python
# find the IQR (inter quantile range) to identify outlines
```

```python
iqr= q3-q1
print('IQR',iqr)
```

```python
# Outlier detection formula
# Higher side ==> Q#+(1.5 * IQR)
# Lower side ==> Q1-(1.5 * IQR)
```

```python
upside = q3 + (1.5 *iqr)
print(upside)
```

```python
lower_side=q1-(1.5*iqr)
print(lower_side)
```
